**1.What is the difference between a function and a method in Python?**

1a - In Python, the key difference between a function and a method lies in their association with objects

1.**function**

**Function**:

A function is a standalone block of reusable code that performs a specific task.
It is defined using the def keyword and can be called independently

```python
#Example
def greet(name):
    return f"Hello, {name}!"

print(greet("pranay"))
```

A method is a function that is associated with an object (usually inside a class).

Methods typically include self as the first parameter to refer to the instance.

**6.Explain the concept of generators in Python and how they are defined?**

6a - **Generators in Python**

A generator is a special type of iterator that allows you to generate values on the fly rather than storing them in memory. It is useful for working with large data sets efficiently.

**1.What Makes a Generator Different?**

Uses the yield keyword instead of return.

Lazy evaluation: Generates values only when needed.

Memory efficient: Doesn't store all values in memory like a list.


**2. Defining a Generator**

A generator function is like a normal function but uses yield to produce a sequence of values.






```python
#Example
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
gen = count_up_to(5)

print(next(gen))
print(next(gen))
print(next(gen))
```

The function does not return all values at once.

Instead, yield pauses the function and resumes on the next call.



**3.Iterating Over a Generator**

Generators are iterators, so they work in loops.

```python
#Example
for num in count_up_to(9):
    print(num)
```

**4.Generator vs. Regular Function**

**Feature  **  	   **Regular Function**	      **Generator**
 Uses               return                     yield
 Memory Usage       Stores all values          Generates one at a time
 Execution          Runs once and exits        Pauses and resumes
 Example            return [1, 2, 3]           yield 1, yield 2, yield 3

**5. Generator Expressions (Like List Comprehensions)**

You can create generators using generator expressions (similar to list comprehensions).

```python
#Example
gen = (x**2 for x in range(5))

print(next(gen))
print(next(gen))
print(next(gen))
```

**7. What are the advantages of using generators over regular functions?**

7a - Advantages of Using Generators Over Regular Functions in Python
Generators provide several benefits over regular functions, especially when working with large datasets, infinite sequences, or memory-intensive tasks.

**1. Memory Efficiency **

Regular functions return lists, which store all values in memory.

Generators generate values one at a time, reducing memory usage.



```python
#Example
def generate_list(n):
    return [i for i in range(n)]

def generate_numbers(n):
    for i in range(n):
        yield i

gen = generate_numbers(10**6)
```

***2. Lazy Evaluation (On-Demand Execution) ***

A generator does not execute all at once.

Values are generated only when requested using next().

```python
#Example
def generate_numbers(n):
    for i in range(n):
        yield i

gen = generate_numbers(5)
print(next(gen))
print(next(gen))
```

**3. Generators Support Infinite Sequences**

Regular functions must return all values, making infinite loops impractical.

Generators pause execution, making infinite sequences possible.

```python
#Example
def infinite_counter():
    num = 1
    while True:
        yield num
        num += 1  # Keeps going

counter = infinite_counter()
print(next(counter))  # Output: 1
print(next(counter))
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

10a - **Difference Between map(), filter(), and reduce() in Python**

These three functions (map(), filter(), and reduce()) are higher-order functions that apply operations to iterables efficiently.

1. map() – Transforms Each Element
Purpose: Applies a function to each item in an iterable and returns a new iterable with the transformed values.

Returns: A map object (iterator), which can be converted to a list.

```python
#Example
numbers = [1, 2, 3, 4]

squared = map(lambda x: x ** 2, numbers)

print(list(squared))
```

2. filter() – Filters Elements Based on a Condition
Purpose: Applies a function that returns True or False and keeps only the elements where the function returns True.

Returns: A filter object (iterator), which can be converted to a list.

```python
#Example: Filtering Even Numbers
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even_numbers = filter(lambda x: x % 2 == 0, numbers)

print(list(even_numbers))
```

**5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?**

5a - Here’s a Python implementation of an iterator class that generates the Fibonacci sequence up to a specified number of terms:


```python
#
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current = 0
        self.next_term = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.num_terms:
            if self.count == 0:
                result = self.current
            else:
                result = self.current
                self.current, self.next_term = self.next_term, self.current + self.next_term
            self.count += 1
            return result
        else:
            raise StopIteration

# Example:
num_terms = 10
fibonacci = FibonacciIterator(num_terms)

for num in fibonacci:
    print(num)
```

**6.Write a generator function in Python that yields the powers of 2 up to a given exponent?**

6a - Here’s a Python generator function that yields the powers of 2 up to a given exponent

```python
#
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example:
exponent = 10
for power in powers_of_two(exponent):
    print(power)
```

10a - You can use the filter() function along with a lambda function to remove all vowels from a given string.

```python
#
def remove_vowels(string):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda char: char not in vowels, string))

# Example:
input_string = "Hello, World! Python is awesome."
result = remove_vowels(input_string)

print("Original String:", input_string)
print("String without Vowels:", result)
```
