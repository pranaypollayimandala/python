6a - Lists are dynamic – they can grow, shrink, and have their elements modified.

9a - Type casting (also called type conversion) in Python is the process of converting one data type into another. Python supports two types of type casting:

Explicit Type Casting (Manual Conversion) – Done using built-in functions.
Implicit Type Casting (Automatic Conversion) – Done automatically by Python.

10a - Type casting (or type conversion) is the process of converting a value from one data type to another. In Python, this can happen in two ways: Implicit and Explicit type casting.

1️ Implicit Type Casting (Automatic)
Done automatically by Python when it is safe.
No data loss occurs during conversion.
Happens when mixing different numeric types (e.g., int and float).

2️ Explicit Type Casting (Manual)
Done manually by the programmer using built-in functions.
Might cause data loss if converting incompatible types.
Common functions used:
int(x) → Converts x to an integer.
float(x) → Converts x to a float.
str(x) → Converts x to a string.
bool(x) → Converts x to a boolean.
list(x), tuple(x), set(x) → Converts x to respective collections.

11a - Purpose of Conditional Statements in Python are Conditional statements in Python are used to control the flow of a program by allowing it to make decisions based on given conditions. They help execute different blocks of code depending on whether a condition is True or False.


12a - The elif (short for "else if") statement is used when you need to check multiple conditions in a program. It allows you to execute different code blocks based on different conditions, instead of writing multiple if-else statements.

13a - Both for and while loops in Python are used to execute a block of code multiple times, but they differ in how they determine the number of iterations

1️ for Loop
Used when the number of iterations is known.
Iterates over a sequence (e.g., list, tuple, string, range).
Automatically stops when the sequence ends.

2️ while Loop
Used when the number of iterations is unknown.
Runs as long as a condition is True.
Requires manual condition update to avoid infinite loops.

14a - A while loop is more suitable when the number of iterations is unknown and depends on a condition being met

example :-1 The number of withdrawals is unknown—it depends on the user.
2 The loop should run until the balance reaches zero, not a fixed number of times.
3 A for loop would require a fixed iteration count, which isn't ideal in this case

```python
#4
import keyword

def is_keyword(word):
    return keyword.iskeyword(word)

word = input("i love my country: ")

if is_keyword(word):
    print(f"'{word}' is a Python keyword.")
else:
    print(f"'{word}' is NOT a Python keyword.")
```

```python
#6
def modify_values(immutable_var, mutable_var):
    immutable_var += 10
    mutable_var.append(10)

    print("Inside function:")
    print("Immutable variable:", immutable_var)
    print("Mutable variable:", mutable_var)

num = 5
numbers = [1, 2, 3]

print("Before function call:")
print("Immutable variable:", num)
print("Mutable variable:", numbers)

modify_values(num, numbers)

print("After function call:")
print("Immutable variable:", num)
print("Mutable variable:", numbers)
```

```python
#7
def arithmetic_operations(a, b):
    print(f"Sum: {a} + {b} = {a + b}")
    print(f"Difference: {a} - {b} = {a - b}")
    print(f"Product: {a} * {b} = {a * b}")

    if b != 0:
        print(f"Quotient: {a} / {b} = {a / b}")
        print(f"Floor Division: {a} // {b} = {a // b}")
        print(f"Remainder: {a} % {b} = {a % b}")
    else:
        print("Division by zero is not allowed.")

    print(f"Exponentiation: {a} ** {b} = {a ** b}")


num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))


arithmetic_operations(num1, num2)
```

```python
#8
def logical_operations(a, b):
    print(f"a = {a}, b = {b}")

    print(f"a > 5 and b > 5: {a > 5 and b > 5}")

    print(f"a > 5 or b > 5: {a > 5 or b > 5}")

    print(f"not(a > 5): {not (a > 5)}")
    print(f"not(b > 5): {not (b > 5)}")

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

logical_operations(a, b)
```

```python
#10
str_list = ["10", "20", "30.5", "40", "True", "False"]

try:
    int_list = [int(item) for item in str_list]
    print("Integer List:", int_list)
except ValueError as e:
    print("Error converting to integers:", e)

try:
    float_list = [float(item) for item in str_list]
    print("Float List:", float_list)
except ValueError as e:
    print("Error converting to floats:", e)

bool_list = [item.lower() == "true" for item in str_list]
print("Boolean List:", bool_list)
```

```python
#14
string = input("Enter a string: ")

reversed_string = ""
index = len(string) - 1
while index >= 0:
    reversed_string += string[index]
    index -= 1

print("Reversed string:", reversed_string)
```

```python
#15
num = int(input("Enter a number: "))

factorial = 1

if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0:
    print("Factorial of 0 is 1.")
else:

    i = num
    while i > 1:
        factorial *= i
        i -= 1

    print(f"Factorial of {num} is {factorial}")
```
