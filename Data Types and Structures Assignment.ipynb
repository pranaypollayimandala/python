**4.Describe how dictionaries store data?**

4a - In Python, dictionaries are hash maps (also known as hash tables) that store key-value pairs efficiently. They allow for fast lookups, insertions, and deletions using a technique called hashing.

1. Internal Structure
A dictionary in Python is implemented using a hash table, which consists of:
Keys: Unique identifiers that are immutable (e.g., strings, numbers, or tuples).
Values: Data associated with each key, which can be of any type.
Hash Function: A function that converts keys into an index where the value is stored.

2. How Data is Stored
Each key is passed through a hash function, generating a unique hash value (integer).
This hash value determines an index (bucket) in memory where the value is stored.
The dictionary uses an array of key-value pairs, where each slot (bucket) can hold a value.

3. How Data is Retrieved
When accessing a value using a key, Python rehashes the key.
It checks the computed hash index and retrieves the value in constant time

4. Collision Handling
Since multiple keys may produce the same hash (a collision), Python uses Open Addressing with Probing:
If a collision occurs, Python searches for the next available slot in the table.
This ensures fast lookups while maintaining efficiency.

5. Dictionary Growth & Resizing
Dictionaries start with a fixed number of slots.
When they reach a certain threshold (e.g., 2/3 full), they resize to a larger size.
The entire table is rebuilt, rehashing and redistributing the key

**8.What is a hash table, and how does it relate to dictionaries in Python?**

8a - A hash table (or hash map) is a data structure that stores key-value pairs and allows for fast lookups, insertions, and deletions. It achieves this efficiency using a hash function, which maps keys to unique indices (buckets) in an array

**How Hash Tables Work**
A key is input into a hash function → Produces a unique hash (integer).
The hash value determines an index in an array where the value is stored.
Lookup, insertion, and deletion are fast because they only require computing the hash and accessing the index.

**Hash Tables and Python Dictionaries**
In Python, dictionaries (dict) are implemented using hash tables. This means:
Dictionary keys are hashed to compute an index in an internal array.
Values are stored at the computed index.
The process enables O(1) average time complexity for lookups and updates

**17.Describe the time complexity of accessing elements in a dictionary?**

17a - Time Complexity of Accessing Elements in a Dictionary (dict)
In Python, dictionaries use a hash table implementation, making element access very efficient.

1. Average Case: O(1) (Constant Time)
Accessing a value by key: dict[key]
Checking membership: key in dict
Python computes the hash of the key and directly retrieves the value from the memory location.
This makes dictionary lookups extremely fast on average.

2. Worst Case: O(n) (Linear Time)
In rare cases, if hash collisions occur (when multiple keys hash to the same slot), Python resolves them by chaining or probing.
If many keys hash to the same slot, lookups degrade to O(n) in the worst case.
However, Python optimizes its hash table to keep performance near O(1) most of the time
